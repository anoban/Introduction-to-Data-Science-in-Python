```python
import numpy as np
import pandas as pd
import re
from datetime import datetime
```

```python
presidents = pd.read_csv(r"D:/Introduction-to-Data-Science-in-Python/week-2/datasets/presidents.csv", index_col = 0)
presidents.head()
```

```python
f_names = presidents.President.apply(lambda full_name: re.findall("[A-Z][\w]+[^\s]", full_name)[0])
f_names
```

```python
l_names = presidents.President.apply(lambda full_name: re.findall("[^\s][A-Z,\.]*[\s]*[a-zA-Z]+", full_name)[1])
l_names
```

```python
presidents.loc[:, "First_name"] = presidents.President.apply(lambda full_name: re.findall("^[A-Z][a-zA-Z]+[^\s]", full_name)[0])
presidents.First_name
```

```python
presidents.loc[:, "Born"] = presidents.Born.apply(lambda dob: re.findall("^[A-Z]{1}[a-z]{2}\s[\d]+\,\s\d{4}", dob)[0])
```

```python
presidents
```

```python
pd.to_datetime(presidents.Born, format = "%b %d, %Y")

# list(map(lambda date_str: datetime.strptime(date_str, "%b %d, %Y"), presidents.Born))
```
