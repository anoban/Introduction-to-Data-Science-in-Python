```python
# multi level indexing
# a data from US census Bureau
census = pd.read_csv(r"D:/Introduction-to-Data-Science-in-Python/week-2/datasets/census.csv")
census.head()
```

```python
# this dataframe contains two sumaries: one for the county and the other for each state

census.SUMLEV.unique()
```

```python
# lets subset the dataframe to consider just the rows with county level summaries i.e SUMLEV = 50

census.query("SUMLEV == 50").head(4)
```

```python
county = census.loc[census.SUMLEV == 50, :]
county.head()
```

```python
county.SUMLEV.unique()
```

```python
# drop all the columns except total number of births & total population estimates

cols_to_keep = ["STNAME", "CTYNAME"] + ["BIRTHS" + str(year) for year in range(2010, 2016)] + \
                                            ["POPESTIMATE" + str(year) for year in range(2010, 2016)]
```

```python
county = county.drop([col_name for col_name in census.columns if col_name not in cols_to_keep], axis = 1, inplace = False)
county.head()
```
