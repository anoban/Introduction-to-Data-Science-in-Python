```python
data = pd.read_csv(r"D:/Introduction-to-Data-Science-in-Python/datasets/house_prices.csv")
data.head()
```

```python
left = data.loc[:, ["street", "city", "zip", "state", "beds", "baths", "sq__ft"]]
left.head()
```

```python
right = data.loc[:, ["sq__ft", "type", "sale_date", "price", "latitude", "longitude"]]
right.head()
```

```python
pd.merge(left, right, on = "sq__ft", copy = False)
```

### ***Inner join:- Joins only common rows from the two tables (an intersect in set terms)***
### ***Outer join:- Joins all rows from both tables, introducing NaNs whenever there is missing data!***
### 

```python
a = pd.DataFrame([{"Name": "Ashley", "Age": 31},
                  {"Name": "Nate", "Age": 35},
                  {"Name": "Graham", "Age": 29},
                  {"Name": "Sarah", "Age": 27}])
a
```

```python
b = pd.DataFrame([{"Name": "Ashley", "Sex": "Female", "ID": 54671226},
                  {"Name": "Nate", "Sex": "Male", "ID": 56765451},
                  {"Name": "Julia", "Sex": "Female", "ID": 86726156},
                  {"Name": "Sarah", "Sex": "Female", "ID": 10105141}])
b
```

```python
# intersect
pd.merge(a, b, how = "inner")
```

```python
# union
pd.merge(a, b, how = "outer")
```

```python
pd.merge(a, b, on = "Name")
```

```python
# joining DataFrames only when multiple columns match

a = pd.DataFrame([{"Name": "Grace", "Study": "Geography", "Age": 26},
              {"Name": "Leslie", "Study": "Mathematics & Statistics", "Age": 25},
              {"Name": "Janice", "Study": "Political Science", "Age": 27},
              {"Name": "Joanne", "Study": "Computer Science", "Age": 31}])
a
```

```python
b = pd.DataFrame([{"Name": "Grace", "Study": "Geography", "Sex": "Female"},
              {"Name": "Leslie", "Study": "Mathematics & Statistics", "Sex": "Female"},
              {"Name": "Janice", "Study": "Political Science", "Sex": "Female"},
              {"Name": "Joanne", "Study": "Biology", "Sex": "Female"}])
b
```

```python
pd.merge(a, b, how = "inner")
```

```python
pd.merge(a, b, how = "outer")
```

```python
pd.merge(a, b, on = ["Name", "Study"])
```
