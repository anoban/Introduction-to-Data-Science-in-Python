```python
def categorizeByStateStartAlphabet(state_name: str) -> str:
    if state_name[0] in ascii_uppercase[:10]:
        return "Category One"
    elif state_name[0] in ascii_uppercase[10:20]:
        return "Category Two"
    elif state_name[0] in ascii_uppercase[20:]:
        return "Category Three"
    else:
        return None
```

```python
_census = census.set_index("STNAME")
```

```python
# group by would operate on index by default if a column name is not specified explicitly

for group, dataframe in _census.groupby(categorizeByStateStartAlphabet):
    # print(group)
    print("{}: States: {}".format(group, list(dataframe.index.unique())))
```

```python
# here we are interested in two columns: the cancellation_policy & review_scores_value

airbnb = pd.read_csv(r"D:/Introduction-to-Data-Science-in-Python/week-3/datasets/listings.csv")
airbnb.head()
```

```python
airbnb.loc[:, ["cancellation_policy", "review_scores_value"]]
```

```python
# group by the first alphabet of the state

groups = dict()

for group, dframe in census.groupby(lambda multiindex: multiindex[0][0]):
    groups[group] = dframe
```

```python
groups.keys()
```

```python
for alphabet in groups.keys():
    print("{}: {}".format(alphabet, {state for state, _ in groups.get(alphabet).index}))
```

```python
airbnb.groupby("cancellation_policy").agg({"review_scores_value": np.mean})
```
