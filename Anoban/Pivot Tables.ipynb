```python
# let's categorize universities into tiers
# using a new column - rank level

# world ranks 1 - 100 first tier
# world ranks 101 - 200 second tier
# world ranks 201 - 300 third tier
# > 300 others

def rank(x: int) -> str:
    if x in range(1,101):
        return "First Tier"
    elif x in range(101, 201):
        return "Second Tier"
    elif x in range(201, 301):
        return "Third Tier"
    elif x in range(301, 1001):
        return "Other"
    else:
        return None

_univ_ranks = pd.Series(data.world_rank.apply(rank), dtype = np.str_, name = "world_tier")
```

```python
data = pd.concat([_univ_ranks, data], axis = 1)
data.head()
```

```python
data.pivot_table(values = "score", index = "country", columns = "world_tier", aggfunc = np.mean)
```

```python
data.pivot_table(index = "country", columns = "world_tier", values = "score", aggfunc = [np.min, np.max])
```
